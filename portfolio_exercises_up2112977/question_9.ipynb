```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image_url = r"/workspaces/m32895_mywork/portfolio_exercises_up2112977/data/fish.png"
image = Image.open(image_url)
image_array = np.array(image)


horizontal_mirror = np.flip(image_array, axis=1)


vertical_mirror = np.flip(image_array, axis=0)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image_array)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(horizontal_mirror)
axes[1].set_title("Horizontal Mirror")
axes[1].axis("off")

axes[2].imshow(vertical_mirror)
axes[2].set_title("Vertical Mirror")
axes[2].axis("off")

plt.show()
```
